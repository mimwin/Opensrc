```python
class Point:
    def __init__(self,first,second):
        self.list=[first,second]
    def show(self):
        print(self.list)
    def __add__(self,other):
        return Point(self.list[0]+other.list[0],self.list[1]+other.list[1])
    
a= Point(1,2)
a.show()
b=Point(3,4)
b.show()
a=a+b
a.show()
```

```python
import csv

f=open('seoul.csv','r',encoding='cp949')
data = csv.reader(f,delimiter=',')

header = next(data)

best = -999
date = 0

for row in data:

    result = 0
    for i in range(0,7):
        if(row[-1]==''):
            row[-1]=-999

        low = float(row[3])
        high = float(row[4])

        result += abs(high-low)
    
    if(result > best):
        date = row[0]

arr = date.split("-")
month = int(arr[1])
day = int(arr[2])
enddate = str(month)+"/"+str(day)
startdate = str(month)+"/"+str(day-6)
        
print("Max Week = "+startdate +" ~ "+enddate)
```
