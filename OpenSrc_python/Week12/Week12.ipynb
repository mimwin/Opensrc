```python
import csv

print("201911155 김미승")
print("실습 #1 : 서울의 일교차 \n ")

f = open('seoul.csv','r',encoding='cp949')
data = csv.reader(f, delimiter=',')

header = next(data)

date = ['','','']  # 날짜
temp = [-999, 999, -999] #일교차, 최저기온, 최고기온

for row in data:
    # row[-1] : row 배열의 가장 마지막 인덱스
    
    # 데이터 누락시
    if row[-1] == '':
        row[-1] = -999
    
    
    low = float(row[3]) # 최저 기온
    high = float(row[4])  # 최고 기온
    
    # 일교차
    if abs(high-low) > temp[0]:
        temp[0] = abs(high-low)
        date[0] = row[0] # 날짜
        
    # 최저 기온
    if low < temp[1]:
        temp[1] = low
        date[1] = row[0]
        
    # 최고 기온
    if high > temp[2]:
        temp[2] = high
        date[2] = row[0]


# 출력
print("일교차가 가장 큰 날")
print(date[0])
print("최저기온이 가장 낮은 날")
print(date[1])
print("최고기온이 가장 높은 날")
print(date[2])

f.close()
```

```python
import csv

print("201911155 김미승")
print("실습 #2 : 지하철 이용객\n ")

f = open('subway.csv','r',encoding='cp949')
data = csv.reader(f, delimiter=',')

headers = next(data) 

slist = []
route = '1호선'

for row in data:
    station = row[1]
    if station == route:
        slist.append((int(row[4].replace(',','')), row[3]))
        
    elif station != route:
        
        slist.sort()
        slist.reverse()
        print("------------------------------")
        print(route)
        print("\n이용객이 가장 많은 역 순위")
        for i in range(3):
            print("{0:d}. {1:s} ({2:d})".format(i+1, slist[i][1], slist[i][0]))
            
        slist.reverse()
        print("\n이용객이 가장 적은 역 순위")
        for i in range(3):
            print("{0:d}. {1:s} ({2:d})".format(i+1, slist[i][1], slist[i][0]))
        print("------------------------------\n")
        route = station
        slist.clear()


f.close()
```
