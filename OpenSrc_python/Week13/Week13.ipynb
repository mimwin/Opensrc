```python
import csv
import matplotlib.pyplot as plt


print("201911155 김미승")
print("실습 #1 : 생일 기온 시각화 \n ")
# 기온 데이터 기간 : 2011-01-01 ~ 2020-12-31 

f = open('seoul_10.csv','r',encoding='cp949')
data = csv.reader(f, delimiter=',')

header = next(data)
max = []
min = []

for row in data:
    month = row[0].split('-')[1]
    day = row[0].split('-')[2]
    if month == '10' and day == '31':
        
        #최고기온
        if row[-1] != '' :
            max.append(float(row[-1]))
        
        #최저기온
        if row[-2] != '' :
            min.append(float(row[-2]))

# 그래프 출력
plt.title('# 2011-01-01 ~ 2020-12-31 #')
plt.plot(max, color='pink', label='max')
plt.plot(min, color='skyblue', label='min')
plt.legend()
plt.show()

f.close()
```

```python
import csv
import matplotlib.pyplot as plt


print("201911155 김미승")
print("실습 #2 : 최고 기온 시각화 \n ")
# 기온 데이터 기간 : 2011-01-01 ~ 2020-12-31 

f = open('seoul_10.csv','r',encoding='cp949')
data = csv.reader(f, delimiter=',')

header = next(data)
one = []
eight = []

for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month == '01':
            one.append(float(row[-1]))
        elif month == '08':
            eight.append(float(row[-1]))

plt.title('# 2011-01-01 ~ 2020-12-31 #')
plt.hist(eight, bins=100, color='r', label='Aug')
plt.hist(one, bins=100, color='b', label='Jan')
plt.legend()
plt.show()

f.close()
```
