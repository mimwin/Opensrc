```python
print(" 실습 #2 : 최대값 구하기\n ")

num = input("Numbers? ")
arr = []
arr = num.split(' ')

for i in range(0,len(arr)):
    arr[i] = int(arr[i])
    
arr.sort()
print("Max = ",arr[len(arr)-1])
```

```python
def crypto(code, flag) :
    arr = []
    n=""

    for i in code: 
        if(i == ' '): 
            arr.append(' ' + n)
            n=""
        else : 
            if(flag==0) : i = chr(ord(i)+1)
            elif(flag==1) : i = chr(ord(i)-1)
            n += i

    arr.append(n)
    arr.reverse()

    n=""
    for i in arr :
        n += i
    
    return n

    
print(" 실습 #3 : 암호화 / 복호화 \n")

code = input("문자열을 입력하시오 : ")

print("암호화 : ",crypto(code,0))

print("복호화 : ",crypto(code,1))
```
