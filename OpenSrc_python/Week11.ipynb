```python
print(" 실습 #6 : copy 구현하기\n ")
print("cp file1 file2 형식으로 입력해주세요.\n")
userinput = input()
filearr = userinput.split(" ")
file1 = open(filearr[1], 'r', encoding='utf-8')
file2 = open(filearr[2], 'w', encoding='utf-8')

while True:
    line = file1.readline()
    if(line == ""): break
    file2.write(line)

file1.close()
file2.close()
print("copy 완료")
```

```python
print(" 실습 #9 : 계산기 구현하기\n ")
class mystack:
    def __init__(self):
        self.stack=[]
    def push(self, a):
        self.stack.append(a)
    def pop(self):
        if not self.stack:
            return -1
        else :
            return self.stack.pop(-1)
    def top(self):
        if not self.stack:
            return -1
        else:
            return self.stack[-1]

num = mystack()
op = mystack()

def calculator():
    inputcal = input("계산식 입력 : ")
    calcarr = inputcal.split(" ")
    
    for i in range(len(calcarr)):
        

        s = calcarr[i]
        
        # 숫자인 경우
        if s.isdigit():    
            
            #숫자 스택에 push
            num.push(int(s))
            
            # 제일 위 연산자 꺼내기
            op1 = op.top()
            
            #연산자 있는 경우
            if op1 != -1:
                
                # 연산자가 *, / 인 경우
                if op1 == "*" or op1 == "/" or (i == len(calcarr) -1):
                    # 연산
                    op1 = op.pop()
                    calc(op1)
                    
                    # 밀린 연산
                    op1 = op.pop()
                    if(op1 != -1):
                       calc(op1)
                        
                
                # 나머지 경우
                elif not (calcarr[i+1] == "*" or calcarr[i+1] == "/"):
                    # 연산
                    op1 = op.pop()
                    if(op1 != -1):
                        calc(op1)
        else:
            op.push(s)
    
    print("계산식 출력:")
    for s in calcarr:
        print(s, end = "")
    print("=", num.pop())
            
def calc(op):
    num2 = num.pop()
    num1 = num.pop()
    if op == "*":
        result =  num1 * num2
    elif op == "/":
        result =  num1 / num2
    elif op == "+":
        result =  num1 + num2
    elif op == "-":
        result =  num1 - num2
    num.push(result)
```
